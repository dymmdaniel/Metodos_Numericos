## Punto 5

Simule un sistema de ecuaciones lineales 111x2134 que contengan números aleatorios enteros.

**Pasos:**

1. Genere una matriz aleatoria A
2. Genere un vector b
3. Resuelva la ecuación $Ax=b$ usando algún paquete de numpy.

Realice un análisis de qué respuestas del sistema tienen componentes mayor de 50 y trunque los resultados redondeando adecuadamente.

Hacer lo missmo con una ecuación **tensorial**


```python
import numpy as np
A = np.random.randint(1,20000,size=(111,2134))#LLenamos una matriz del tamaño 111*2134 de numero aleatorios del 1 al 20000
b = np.random.randint(1,20000,size=(111,1))#LLenamos un vector del tamaño 111 de numero aleatorios del 1 al 20000
x= np.linalg.lstsq(A, b,rcond=-1)[0] #Devuelve la solución de mínimos cuadrados a una ecuación de matriz lineal.
x

```

```python
resultado=np.rint(x[x>50])
if len(resultado)==0:
    print("Ups, no hay nada, intenta un numero mas pequeño :D")
else:
    print(resultado)
```
